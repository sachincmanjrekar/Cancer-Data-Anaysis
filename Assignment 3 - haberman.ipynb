# Objective:
    1.Determining life status of random person who lived in between 50's and 60's based on his age and no. of cancer nodes that he had.

### 1D-Scatter plot

```python
import numpy as np
lived=hb.loc[hb["status"]==1]
dead=hb.loc[hb["status"]==2]
plt.plot(lived["nodes"],np.zeros_like(lived['age']),'o',label="lived")
plt.plot(dead["nodes"],np.zeros_like(dead['age']),'o',label="dead")
plt.title("1D scatter plot considering number of nodes")
plt.xlabel("No. of Nodes")
plt.legend(loc="best")
plt.show()
```

###### Conclusion using 1D scatter plot:
    
    1.Cannot make any observation out of 1D scatter plot because of massive overlapping.

### 2D-Scatter plot

```python
#total combination 3C2=3
#1.nodes&year 2.age&year 3.age&nodes
 
hb.plot(kind='scatter',x='nodes',y='year');
plt.title("2D Scatter plot")
plt.show()
```

### PAIR PLOT

```python
plt.close();
sn.set_style("whitegrid");
sn.pairplot(hb,vars=["age","year","nodes"], hue="status", size=3);
plt.show()
```

### Box Plot

```python
sn.boxplot(x='status',y='nodes',data=hb)
plt.title("Box plot")
plt.show();
```

###### Observations using Box plots:
    
    Box plots are bit useful.because if a person has nodes lets say more than 4 he had less chance of surviving for atleast 5 years.

### Violin plot

```python
sn.violinplot(x="status",y="nodes",data=hb,size=8) 
plt.title("violin plot");
plt.show();
```

###### Observations using Violin plots:

    Violinplots are not useful towards classification if we use "age" and "year", but if we use "nodes" factor we can see some positives that person who has less no. nodes has better chance of surviving.

### HISTOGRAM

```python
#univariate analysis

sn.FacetGrid(hb, hue="status",height=5).map(sn.distplot,"nodes").add_legend();
plt.title("life status of a person considering the no. of nodes they have");
plt.show();
```

```python
sn.FacetGrid(hb, hue="status",size=5).map(sn.distplot,"year").add_legend();
plt.title("life status of a person considering their decease detection year");
plt.show();
```

```python
sn.FacetGrid(hb, hue="status",size=5).map(sn.distplot,"age").add_legend();
plt.title("life status of a person considering age");
plt.show();
```

### Conclusion using Univariate analysis:
    
    1.If we use "nodes" factor, we can conclude that person who has very few nodes(less than 4) has better chance of surviving.
    2."age" and "the year decease got detected" doesn't affect thier survival status.
    3.In simple words there wasn't much medical breakthrough in 50's and 60's. People who have "higher cancer nodes" more likely to die

## PDF,CDF

```python
counts, bin_edges = np.histogram(lived['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute PDF,CDF of patients according to their age
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="pdf")
plt.plot(bin_edges[1:],cdf,label="cdf")
plt.xlabel("The year of detection")
plt.title("People who lived acccording to The Year nodes detected ")
plt.legend(loc="best");
plt.show();
```

```python
#pdf,cdf of life using number of nodes
counts, bin_edges = np.histogram(lived['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="pdf of lived")
plt.plot(bin_edges[1:],cdf,label="cdf of lived")
plt.legend(loc="best")


#pdf,cdf of death using number of nodes
counts, bin_edges = np.histogram(dead['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="pdf of dead ") 
plt.plot(bin_edges[1:], cdf,label="cdf of dead")
plt.legend(loc="best")
plt.xlabel("No. of nodes")
plt.title(" acccording to The number of Nodes detected ")
plt.show();
```
